```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Sample User-Item Matrix
data = {
    "Movie":["Movie1","Movie2","Movie3","Movie4"],
    "User1":[5,3,0,1],
    "User2":[4,0,0,1],
    "User3":[1,1,0,5],
    "User4":[0,0,5,4]
}

df = pd.DataFrame(data)
df.set_index("Movie", inplace=True)

# Compute Similarity
similarity = cosine_similarity(df)

similarity_df = pd.DataFrame(
    similarity,
    index=df.index,
    columns=df.index
)

print("Movie Similarity Matrix")
print(similarity_df)

# Recommendation Function
def recommend(movie):
    similar_movies = similarity_df[movie].sort_values(ascending=False)
    return similar_movies[1:3]

print("\nRecommendations for Movie1:")
print(recommend("Movie1"))
```
